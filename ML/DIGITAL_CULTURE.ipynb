```python
import numpy as np 



X = np.array([13, 4, 11, 20])
Y = np.array([8, 5, 6, 15])


X_hat = np.sum(X)/len(X)
Y_hat = np.sum(Y)/len(Y)
```

```python
teta_1 = np.sum((X - X_hat) * (Y - Y_hat))/np.sum((X - X_hat)**2) 
```

```python
round(teta_1, 3)
```

```python
teta_0 = Y_hat - X_hat * teta_1
```

```python
round(teta_0, 3)
```

```python
n = len(X)
```

```python
SE_teta0 = (np.sum((Y - teta_0 - teta_1 * X)**2) / (n-2))**0.5 * (1/n + X_hat**2 / np.sum((X - X_hat)**2))**0.5
```

```python
round(SE_teta0, 3)
```

```python
SE_teta1 = (((np.sum((Y - teta_0 - teta_1 * X)**2)) / (n-2))**0.5) * (1/(np.sum((X - X_hat)**2)))**0.5
```

```python
round(SE_teta1, 3)
```

```python
teta0_minus = teta_0 - t_975 * SE_teta0
teta0_plus = teta_0 + t_975 * SE_teta0

teta1_minus = teta_1 - t_975 * SE_teta1
teta1_plus = teta_1 + t_975 * SE_teta1
```

```python
def estimate_coef(x, y): 
    n = np.size(x) 
    m_x, m_y = np.mean(x), np.mean(y)
    print(f'выборочное среднее x_hat:{m_x}, выборочное среднее y_hat:{m_y}') 

    SS_xy = np.sum(y*x) - n*m_y*m_x 
    SS_xx = np.sum(x*x) - n*m_x*m_x 

    theta_1 = SS_xy / SS_xx 
    theta_0 = m_y - theta_1*m_x 

    return(theta_0, theta_1) 
```

```python
import numpy as np
import pandas as pd
from scipy.spatial import distance as dist
from sklearn.cluster import KMeans
```

```python
# чтение данных
data = pd.read_csv("/content/kmeans.csv", delimiter=',', index_col='Object')
```

```python
# удаление столбца Cluster из набора данных
coords = data.drop('Cluster', axis=1)
```

```python
# координаты центроидов
centroid = np.array([[8.0, 12.0], [10.57, 7.43], [9.5, 9.5]])
```

```python
# инициализация модели
k_means = KMeans(n_clusters=3, init=centroid, max_iter=100, n_init=1)
```
